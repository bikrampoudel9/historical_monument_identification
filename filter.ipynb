```python
import pandas as pd

df_labels = pd.read_csv('train_label_to_category.csv')
```

```python
nepal_keywords = [
    "Nepal", "Pashupatinath","Swayambhunath","Boudhanath","Baudhanath","Kathmandu",
    "Patan","Bhaktapur","Hanuman","Changu_Narayan","Lumbini","Maya_Devi","Janaki_Mandir",
    "Muktinath","Manakamana","Nyatapola","Krishna_Mandir","Kasthamandap","Dharahara","Bhimsen_Tower","Gorkha_Durbar","Nuwakot_Durbar""Rani_Mahal",
    "Budhanilkantha","Ashoka_Pillar","Shivapuri","Sagarmatha","Lalitpur"
]
```

```python
# Separate only Nepal landmarks
pattern = '|'.join(nepal_keywords)
nepal_landmarks = df_labels[df_labels['category'].str.contains(pattern, case=False, na=False)]

print(f"Found {len(nepal_landmarks)} unique Nepal monuments.")
```

```python
nepal_landmarks.to_csv('nepal_monuments.csv', index=False)
```
